```python
REPO_PATH = 'C:\\JosephYu\\serverless-python-demo'

import os

def print_directory_contents(path, indent_level=0):
    # Create the indentation string
    indent = ' ' * 4 * indent_level
    # Determine the type of current path (file or folder)
    if os.path.isdir(path):
        print(f'{indent}[L-{indent_level}:ðŸ“‚FOLDER]{os.path.basename(path)}/')
        # Recursively call the function for each item in the directory
        for item in sorted(os.listdir(path)):
            print_directory_contents(os.path.join(path, item), indent_level + 1)
    else:
        print(f'{indent}[L-{indent_level}:ðŸ“‘file]{os.path.basename(path)}')

# Replace 'your_repo_path' with the path to your local repository
print_directory_contents(REPO_PATH)
```
